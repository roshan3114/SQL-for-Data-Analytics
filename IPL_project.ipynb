```python
#get all the column names from dataset 
cols = auc23.columns.to_list()
cols.remove('Season')
```

```python
#remove leading and traling space from every data
for col  in cols: 
  auc23[col] = auc23[col].str.strip()


#convert price to numerical value 
auc23['Price'] = auc23['Price'].str.replace(',','_').astype(np.float64)
```

```python
#Look for number of unique values in each column with object values 
for col in cols:
  print (f"unique values in column {col}is {auc23[col].nunique()}",end='\n')
```

```python
auc23['Type'].value_counts()
```

#### Usually there are 4 category of roles in Cricket. Lets fix that repeating values

```python
# same meaning of Batter and batsman 
# same as weeket keeper and wicket-keeper
# lets change it 
auc23['Type'] = auc23['Type'].str.replace('-',' ')
auc23['Type'] = auc23['Type'].str.replace('Batter', 'Batsman')
```

```python
fig1= px.box(
          auc23,
          y='Price',
          x='Type',
          color='Type',
          title = 'player Type VS Price'
          )
fig1.show()
```

### Overview
* **25%** of the `Batsman`s cost less than **4.5M** and **75%** of them cost less than **39M**.  
* **25%** of the `All Rounders`s cost less than **2M** and **75%** of them cost less than **11.25M**.
* **25%** of the `Bowlers`s cost less than **3.75M** and **75%** of them cost less than **13.75M**.
* **25%** of the `Weeket Keepers`s cost less than **2M** and **75%** of them cost less than **16M**.

### Number Of Players Signed by Team 
`Higher number of players signed means:`
* higher number of vacant place
* Teams want some better players at some role 

```python
fig = px.histogram(auc23,
                   x='Team',
                   color='Type',
                   barmode='group',
                   hover_data = ['Type', 'Team']
                   )
fig.layout=dict(
    title='<b>Numbers pf players signed by teams<b>',
    xaxis=dict(title = '<b>Teams</b>'),
    yaxis=dict(title='<b>Number of Players signed</b>'),
)
fig.show()
```

### Overview 
* punjab kings has the highest all rounders 
* rajsthan royals has the highest bowler's
* mumbai indians has 14th all rounder and 9 bowlers

### Team Exoenditure by Role 
`Team spend their highest amount to fill in the vacant roles.`
`The amount may be higher due to:`
* High Bid Amount

```python
fig2= px.histogram(auc23,
                   y='Price',
                   x='Team',
                   color='Type',
                   barmode='group',
                   hover_data=['Type','Team']
                   )

fig2.layout=dict(
    title='<b>Amount spent by teams to sign players<b>',
    xaxis=dict(title='<b>Teams</b>'),
    yaxis=dict(title='<b>Amount Spent</b>'),
)
fig2.show()
```

```python
fig=px.funnel(auc23,
              x=sorted(auc23.groupby('Team')['Price'].sum()),
              y=auc23['Team'].unique(),
              )
fig.layout=dict(
    title='<b>Expenditure On Players By Team<b>',
    xaxis=dict(title='<b>Expenses</b>'),
    yaxis=dict(title='<b>Team</b>'),
)
fig.show()
```
